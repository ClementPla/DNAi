```python
%load_ext autoreload
%autoreload 2

from scipy import stats
import numpy as np
import pandas as pd
import plotly.express as px

import statsmodels.formula.api as smf

from dnafiber.data.intergrader.plot import add_p_value_annotation
import catppuccin
pd.options.mode.copy_on_write = True
```

```python
df_raw = pd.read_pickle("properties_with_IA.pkl")

categories = df_raw['image_type'].unique()
df_raw['annotator'] = df_raw['annotator'].astype('category')
df_raw['annotator'] = df_raw['annotator'].cat.reorder_categories(['H', 'S', 'Y', 'C', df_raw['annotator'].astype('category').unique()[-1]], ordered=False)
```

```python
for category in categories:

    df = df_raw[df_raw['image_type'] == category]

    # df = df[df.common == 5]
    # df = df[df.ratio <3]
    filtered = df[(np.isin(df.fiber_type, ["double"]))]
    try:
        model = smf.mixedlm("ratio ~ annotator", filtered, groups=filtered["image_type"])
    except ValueError:
        continue
    try:
        result = model.fit()
    except np.linalg.LinAlgError:
        print("LinAlgError")
        continue
    pvalues = np.asarray(result.pvalues)[1:]



    filtered['Ratio'] = filtered['ratio'].astype(float)
    filtered['Annotator'] = filtered['annotator']

    fig = px.violin(
        filtered,
        y="Ratio",
        x="Annotator",
        box=True,  # draw box plot inside the violin
        points="all",  # can be 'outliers', or False
        color="Annotator",
        template="ggplot2",
        title=f"{category}",
    )
    fig = add_p_value_annotation(fig, [[0,1], [1,2], [1,3], [1, 4]], stats_test=pvalues)
    # Set title to the name of the category
    # Increase top margin to make room for the title
    fig.update_layout(
    title=dict(font=dict(size=15), automargin=True, yref='paper')
)
    fig.update_layout(margin=dict(t=115))
    fig.show()
```

```python

colors = [c.hex for c in catppuccin.PALETTE.latte.colors]
filtered['Type'] = filtered['image_type']
groupped = filtered.groupby(['image_type', 'Annotator']).median(numeric_only=True)
for i, group in enumerate(groupped.index):
    if group[1] == 'H':
        filtered.loc[(filtered['image_type'] == group[0]), 'Median'] = groupped.iloc[i]['Ratio']

filtered.sort_values(by=['Median', 'Annotator'], ascending=True, inplace=True)

fig = px.box(
    filtered,
    y="Ratio",
    x="Annotator",
    points=False,  # can be 'outliers', or False
    color="Type",
    template="ggplot2",
    color_discrete_sequence=colors,)
# fig = add_p_value_annotation(fig, [[0,1], [1,2], [1,3]], stats_test=stats.mannwhitneyu)

# Increase top margin to make room for the title
fig.update_layout(margin=dict(t=75))
fig.show()
```
