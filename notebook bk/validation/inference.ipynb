```python
%load_ext autoreload
%autoreload 2
from pathlib import Path
from dnafiber.ui.utils import get_image_cacheless, get_model_cacheless, MODELS_ZOO
from dnafiber.ui.inference import inference_cacheless
from dnafiber.data.intergrader.analysis import (
    extract_properties_from_datadict,
    skeletonize_data_dict,
)
import numpy as np
from tqdm.auto import tqdm
import torch
```

```python
modelname = list(MODELS_ZOO.keys())[1]
modelid = MODELS_ZOO[modelname]
device = "cuda"
model = get_model_cacheless(device, modelid)

list_files = list(Path("C:/Users/cleme/OneDrive/PostDoc/data/DNAFiber/Validation/").rglob("*.czi"))
```

```python
output = dict()
output[modelname] = dict()
for i, file in enumerate(tqdm(list_files)):
    img = get_image_cacheless(file)
    results = inference_cacheless(model, img, device).astype(np.uint8)
    name = file.stem.split('-')[0]
    if name not in output[modelname]:
        output[modelname][name] = []
    output[modelname][name].append(results)
    torch.cuda.empty_cache()    
```

```python
skeletons = skeletonize_data_dict(output)
properties = extract_properties_from_datadict(skeletons, with_common_analysis=False)
```

```python
import pandas as pd

df = pd.DataFrame.from_dict(properties)
df.to_pickle(f"{modelname}_properties.pkl")
```
