```python
import pandas as pd
import plotly.express as px

df = pd.read_pickle("segformer MiT-B2_properties.csv")
df = df[(df['ratio'] < 5) & (df['ratio'] > 0.25)]
df = df[(df['green'] + df['red']) > 225]
df = df[df['fiber_type'] == 'double']
df['Ratio'] = 1 / df['ratio']
df['Type'] = df['image_type'].apply(lambda x: x.replace(" KO", "").replace("SMARCAL", "SMARCALKO").replace("siBRCA2+", "siB+"))
df["Annotator"] = df["annotator"]
xl = pd.ExcelFile("c:/Users/cleme/OneDrive/PostDoc/data/DNAFiber/Validation/YM-2024-59.xlsx")
sheet_names = xl.sheet_names  # see all sheet names
summary_data = dict(Type=[], Ratio=[], Annotator=[])
for sheet in sheet_names:
    data = xl.parse(sheet, header=None)  # read a sheet into a DataFrame
    ratio = data.iloc[:, 2].dropna().values
    summary_data['Ratio'].extend(ratio)
    summary_data['Type'].extend([sheet] * len(ratio))
    summary_data['Annotator'].extend(["Yosra"] * len(ratio))

df_yosra = pd.DataFrame(summary_data)

df_ai = df[['Ratio', 'Type', "Annotator"]].copy()
cat = pd.concat([df_yosra, df_ai], ignore_index=True)

fig = px.violin(
    cat,
    y="Ratio",
    x="Type",
    box=True,  # draw box plot inside the violin
    points="all",  # can be 'outliers', or False
    color="Annotator",
    template="ggplot2",
)
fig.show()
```
